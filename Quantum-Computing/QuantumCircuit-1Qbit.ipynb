```python
from qiskit import *
import numpy as np
n = 1 
qr = QuantumRegister(n) # κβαντικός καταχωρητής
cr = ClassicalRegister(n) # κλασσικός καταχωρητής
qc = QuantumCircuit(qr,cr) # κβαντικό κύκλωμα
qc.initialize([1/np.sqrt(2), -1/np.sqrt(2)], 0) # αλλαγή κατάστασης στο 0 qubit
qc.measure(qr,cr) # μέτρηση qubit
qc.draw(output='mpl') # εμφάνιση κυκλώματος
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(qc, backend = simulator).result()
from qiskit.tools.visualization import plot_histogram
plot_histogram(result.get_counts(qc))
```
